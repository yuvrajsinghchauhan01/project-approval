```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# load data
raw_df = pd.read_csv('/Users/yuvraj/project-approval/Dataset/train_data.csv')
raw_df.head()
```

```python
# null values
raw_df.isnull().sum()/raw_df.shape[0]
```

```python
# countplot for teacher_prefix
sns.countplot(x='teacher_prefix', data=raw_df,hue='teacher_prefix',order=raw_df['teacher_prefix'].value_counts().index)
plt.title('Teacher Prefix')
plt.show()
```

```python
# teacher prefix vs project is approved
sns.countplot(x='teacher_prefix', hue='project_is_approved',data=raw_df,order=raw_df['teacher_prefix'].value_counts().index)
plt.title('Which teacher prefix has most approved projects')
plt.show()
```

```python
# countplot Top 10 school states
sns.countplot(x='school_state', data=raw_df,order=raw_df['school_state'].value_counts().index[:10])
plt.title('Top 10 School States')
plt.show()
```

```python
# countplot Top 10 school states vs project is approved
sns.countplot(x='school_state', hue='project_is_approved',data=raw_df,order=raw_df['school_state'].value_counts().index[:10])
plt.title('Which school state has most approved projects')
plt.show()
```

```python
# countplot project grade category
sns.countplot(x='project_grade_category', data=raw_df,order=raw_df['project_grade_category'].value_counts().index)
plt.title('Project Grade Category')
plt.show()
```

```python
# countplot project grade category vs project is approved
sns.countplot(x='project_grade_category', hue='project_is_approved',data=raw_df,order=raw_df['project_grade_category'].value_counts().index)
plt.title('For which grade project is approved most')
plt.show()
```

```python
# countplot for top 10 project subject categories
sns.countplot(y='project_subject_categories', data=raw_df,order=raw_df['project_subject_categories'].value_counts().index[:10])
plt.title('Top 10 Project subject categories')
plt.show()
```

```python
# countplot for top 10 project subject categories vs project is approved
sns.countplot(y='project_subject_categories', hue='project_is_approved',data=raw_df,order=raw_df['project_subject_categories'].value_counts().index[:10])
plt.title('For which subject project is approved most')
plt.show()
```
